```python
from datasets import Dataset, load_dataset
import ast, astor
def extract_assert_statements(code_snippet):
    tree = ast.parse(code_snippet)
    assert_statements = []
    for node in ast.walk(tree):
        if isinstance(node, ast.Assert):
            assert_source = astor.to_source(node)
            if assert_source.find("assert") != -1:
                assert_statements.append(assert_source)
    return assert_statements

dataset = load_dataset("openai_humaneval", split='test')
[extract_assert_statements(d['test']) for d in dataset]
```

```python
import json
for d in dataset:
    # file_name = d['task_id'].replace('/')
    with open(f"./dataset/{d['task_id']}--.txt", 'w') as f:
        f.write(d['prompt'])
```
